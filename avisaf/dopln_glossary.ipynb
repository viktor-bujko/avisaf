```python
import json
from pathlib import Path
```

```python
news = []
file_name = "../zmazat.txt" 
with Path(file_name).open("r") as news_f:
    news = news_f.readlines()
len(news)
```

```python
news[0].strip().split("\t")
```

```python
with Path("../config/aviation_glossary.json").open("r") as vocab_f:
    vocab = json.load(vocab_f)
```

```python
news_dict = {}
for new in news:
    x = new.strip().split("\t", maxsplit=1)
    key, value = x[0], x[1]
    news_dict[key] = value
```

```python
len(news_dict)
```

```python
len(vocab)
```

```python
new_vocab = vocab.copy()

for idx, (vocab_key, expl_dict) in enumerate(vocab.items()):
    print(idx / len(vocab.items()))
    full = expl_dict.get("full") 
    if full and full != "n/a":
        continue

    new_expl_dict = expl_dict.copy()
    new_expl = news_dict.get(vocab_key)
    if new_expl:
        new_expl_dict["full"] = new_expl
        new_vocab[vocab_key] = new_expl_dict

with Path("../config/aviation_glossary_1.json").open("x") as new_file:
    json.dump(new_vocab, new_file, indent=4)
```

```python
new_vocab = vocab.copy()
for key, value in news_dict.items():
    if vocab.get(key) is not None:
        key_dict = vocab.get(key)
        z = key_dict.get("full") 
        if z and z == "n/a":
            new_key_dict = key_dict.copy()
            new_key_dict["full"] = value
            new_vocab.update({key: new_key_dict})

len(new_vocab.keys())
input("ENTER")
with Path("../config/aviation_glossary_1.json").open("x") as n:
    json.dump(new_vocab, n)
```
